# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
# Prob_2: Take another example and write the code that runs in fixed amount of time or has fixed number of steps of execution no matter what is the size of input has constant time complexity.
# similar to example-1, also plot the graph (eg. even_odd )

import matplotlib.pyplot as plt

def is_even_or_odd(number):
    # This operation takes a constant amount of time, regardless of the 'number' value.
    return 'Even' if number % 2 == 0 else 'Odd'

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    # Simulate a constant time operation. The actual time doesn't change with 'n'.
    is_even_or_odd(n)
    time_values.append(1) # Represent constant time as 1 unit

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Even/Odd Check")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)

import matplotlib.pyplot as plt

def count_positives(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    # Create a list of 'n' numbers for simulation
    data = list(range(-n // 2, n // 2 + 1)) # Example: for n=10, data might be [-5,-4,...,4,5]
    count_positives(data)
    time_values.append(n) # Linear time complexity is represented by 'n'

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for Counting Positives")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithmic time. plot graph as well similar to above example.

import matplotlib.pyplot as plt
import math

def binary_search_iterations(arr, target):
    low = 0
    high = len(arr) - 1
    iterations = 0
    while low <= high:
        iterations += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return iterations
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return iterations


n_values = list(range(1, 1001, 10))  # Input sizes from 1 to 1000
time_values = []

for n in n_values:
    data = list(range(n)) # Create a sorted list of size n
    # For binary search, the number of iterations is proportional to log(n)
    # We'll simulate finding an element (e.g., the last element or middle element)
    # The worst case for binary search is log2(n) iterations.
    if n > 0:
        time_values.append(math.log2(n))
    else:
        time_values.append(0)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity (simulated)")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.

import matplotlib.pyplot as plt
import math

def log_linear_operation(n):
    # Simulate n log n complexity
    # For simplicity, we'll return n * log2(n) as the 'time' for an input size 'n'
    if n <= 1:
        return 0
    return n * math.log2(n)

n_values = list(range(1, 101)) # Input sizes from 1 to 100
time_values = []

for n in n_values:
    time_values.append(log_linear_operation(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (simulated)")
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in polynomial time. plot the graph as well, similar to above example

import matplotlib.pyplot as plt

def polynomial_time_pairs(n):
    # Simulate n^2 complexity
    # This function iterates through all possible pairs in a list of size n
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count # The number of operations is n*n = n^2

n_values = list(range(1, 101)) # Input sizes from 1 to 100
time_values = []

for n in n_values:
    time_values.append(polynomial_time_pairs(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n^2)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in exponential time. plot the graph as well, similar to above example

import matplotlib.pyplot as plt

def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

def count_fibonacci_calls(n, memo={}):
    if n <= 1:
        return 1
    if n in memo:
        return memo[n]
    # Simulate the number of calls for a recursive fibonacci calculation
    # This is not the exact number of operations, but proportional to the exponential growth
    memo[n] = count_fibonacci_calls(n-1, memo) + count_fibonacci_calls(n-2, memo) + 1
    return memo[n]

n_values = list(range(1, 20)) # Exponential growth means we can only go up to small n
time_values = []

for n in n_values:
    time_values.append(count_fibonacci_calls(n)) # Using a helper to count operations

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (c^n)")
plt.title("O(c^n) Exponential Time Complexity (Recursive Fibonacci)")
plt.show()
```

```python
# Write a function to perform quick sort and show its time complexity (O(n log n) average case).
import matplotlib.pyplot as plt
import random
import time
import sys

sys.setrecursionlimit(2000)

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

def measure_quicksort_time(n):
    data = [random.randint(0, n * 10) for _ in range(n)]
    start_time = time.time()
    quicksort(data)
    end_time = time.time()
    return (end_time - start_time) * 1000
n_values = list(range(10, 1001, 50))
time_values = []

for n in n_values:
    num_trials = 5
    avg_time = sum(measure_quicksort_time(n) for _ in range(num_trials)) / num_trials
    time_values.append(avg_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Quick Sort (Observed Time)')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (ms)")
plt.title("Quick Sort Time Complexity (Average Case O(n log n))")
plt.legend()
plt.grid(True)
plt.show()
```
